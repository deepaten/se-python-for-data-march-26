# Data cleaning with pandas

```python
import numpy as np
import pandas as pd
```

```python
# import messy data
df = pd.read_csv("students_data_messy.csv")
print(df.head())
```

### step 1 - Dealing with missing values


- Option 1 - Drop rows/columns contains missing values
- Option 2 - Fill missing values with some values (static, avg)

## drop missing values

```python
# Drop any row that has at least 1 missing value
df_dropped = df.dropna()
print(df_dropped.head())
```

```python
df_dropped_col = df.dropna(subset=["Age"])
print(df_dropped_col.head(10))
```

### Filling missing values

```python
# Fill missing age values with static number
df["Age Filled"] = df["Age"].fillna(0)
print(df.head(5))
```

```python
# Filling missing vlaue with avg, mean median
df["Score Filled"] = df["Score"].fillna(df["Score"].mean())
print(df.head(20))
```

### step 2 Filtering helps explore subset of data with specific conditions
### sorting helps quickly finding highest/lowest values

```python
# make df for age over 30
df_17_over = df[df["Age"] > 17]
print(df_17_over)
```

```python
# Filter for multiple conditions
high_score = df[(df["Age"] > 17) & (df["Score"] > 90)]
high_score.head()
```

```python
# Sorting by multiple columns
df_sorted_multi = df.sort_values(by=["Age", "Score"], ascending=[True, False])
df_sorted_multi.head(15)
```

# Renaming columns

```python
# spec column
df_renamed = df.rename(columns={"Age": "Age_years", "Score": "Test_Score"})
df_renamed.head()
```
