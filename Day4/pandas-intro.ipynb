# Intro to pandas

```python
import numpy as np
import pandas as pd
```

- Pandas is a Python Library for Data manipulation and analysis
- Built on top of NumPy
- Series = 1D -> Data Structure --> like a column
- Dataframe = 2D -> Data Structure --> like a spreadsheet or sql table etc
- Always explore data first before doing analysis (shape, row, column)
- [Pandas documentation](https://pandas.pydata.org/docs/user_guide/index.html#user-guide)

# Series

# Creating a series

```python
data = [10, 20,30, 40 ,50]
series = pd.Series(data)
print(series)
```

```python
# Accessing values and index
print(series.values)
print(series.index)
```

```python
temps = pd.Series([12, 18, 19 ,25])
print("Max(temps:)", temps.max())
print("Min(temps:)", temps.min())

```

Notes - 
- Series behave like a list but has an index


# Dataframes

```python
# Creating dataframes

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['London', 'Paris', 'Berlin', 'Madrid']}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
data1 = {
   "Product": ["Book", "Pen", "Notebook"],
   "Price": [10, 2, 5],
   "Quantity": [3, 10, 7]}

df1 = pd.DataFrame(data1)
print(df1) # first part
# Add a new column for total value

df1["Total Value"] = df1["Price"] * df1["Quantity"]
print(df1)
```

## loc() and iloc()

- loc() uses names to get the locations of values
- iloc() uses numbers to get the locations

```python
students_name = students_df["Name"]
print(students_name)
```

```python
students_name_loc = students_df.loc[0:2,"Name":"Grade"]
print(students_name_loc)
```

```python
print(students_df.sort_values(by="Score",ascending=False))
```

```python
print(students_df[students_df["Score"]>=80])
```

```python
# Filter by numeric value, sort ascending
print(students_df[students_df["Score"] >= 80].sort_values(by="Score", ascending=False))
```

```python
# Average score by school
print(students_df.groupby("School")["Score"].mean())
```

```python
# All students from a specific school
print(students_df.loc[:,"School"] )
```

```python
# All students from a specifc school
print(students_df.loc[students_df["School"] == "West High"])
```

```python
# Scores of the last 2 students
print(students_df.iloc[-2:, students_df.columns.get_loc("Score")])
```

```python
# Filter students with scores above 80
print(students_df[students_df["Score"] > 80])
```
