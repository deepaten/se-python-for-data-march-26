## Types of Visualisation

- There are many types of Visualisation we can create with Seaborn, We will look at some common types in this notebook

## Scatterplot

```python
sns.scatterplot(x="tip" , y="total_bill" , data=tips, hue = "day", )
```

```python
# Add size to the graph 
sns.scatterplot(x="tip" , y="total_bill" , data=tips, hue = "day", size="size")
```

```python
# Add size to each# Add hue parameter to make colour of points better
sns.scatterplot(x="tip" , y="total_bill" , data=tips, hue="day", size="size", palette="YlGnBu")
```

- `sns` is the standard alias for Seaborn
- Scatterplot() is a Seaborn function for quick and easy Scatterplot

| Parameter | What it does |
| ------------------ | ----------------------------------------------------------------------------------------------------------------------------- |
| `x='tip'` | Tells Seaborn which column of your dataset to plot on the x-axis (`tip` column). |
| `y='total_bill'` | Column for the y-axis (`total_bill`). |
| `data=tips` | The DataFrame (often a pandas `DataFrame`) containing your data. Here it’s the famous `tips` dataset that comes with Seaborn. |
| `hue='day'` | Groups points by the **day** column. Each day gets a different color. |
| `size='size'` | Varies the **marker size** according to the `size` column. Larger party sizes → bigger points. |
| `palette='YlGnBu'` | Chooses the color palette. `'YlGnBu'` (“Yellow-Green-Blue”) is a built-in sequential palette from Matplotlib. |

### Histplots

What are Histograms and distplots?
- Histograms group data into bins and shows frequency distribution
- Distribution plot is a smoothed curve that represents probability distribution
Why are they useful?
- Understand distribution and spread of a SINGLE variable
- Detect skews, peaks etc.
- Example: Checking whether exam marks follow a normal distribution

```python
sns.histplot(tips["tip"], kde = True, bins = 20)
```

```python
# Barplot/ Barchart
```

```python
sns.barplot(x="sex", y="tip", data=tips, hue="sex", palette="mako")
```

# Boxplot

A Boxplot summarises distribution of a dataset by showing the median, quartiles and outliers
Why use one?:
- Good for comparing distributions between different groups and spotting outliers
 Example: Comparing exam results between classes

```python
# Distribution of tips across different days

sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="mako")
```

# Stripplot

A stripplot displays individual data points, often layered with a box plot
Why use one?:
- Adds detail by showing the spread of actual observations

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="mako")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="mako", dodge=True)
```

Join Plot = A combined plot showing both a scatter plot of two variables and their individual distributions around the margins (i.e histplot)
Why use a Join Plot?
- The analyse bivariate relationships while also examing distribution
Pair Plot = Produces scatterplots for every possible pair of variable in your dataframe! Very cool
Why use a Pair Plot?
- To explore across multiple variables simultaneously
- Really useful for early/initial data exploreation

```python
# Joinplot
sns.jointplot(x="tip", y="total_bill", data=tips)
```

```python
# Pairplot
sns.pairplot(data=tips)
```

```python
# basic linechart 
sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
# Add title and change the label
sns.lineplot(x="timepoint", y="signal", data=fmri)

# Use matplotlib to change labels and title of chart
plt.xlabel("Timepoint")
plt.ylabel("Percentage signal change")
plt.title("FMRI signal across time")
```

```python
# Add hue to add individual line to each event
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", palette=["teal", "indigo"])
```
