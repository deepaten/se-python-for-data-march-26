# Load Gold dataset and perform statistical calculations

```python
import csv
```

```python
rows = []
with open("gold_price_data.csv",newline="") as csvfile:
    csvreader = csv.DictReader(csvfile)
    for row in csvreader:
        rows.append(row)
       
close_arr = []

for line in rows:
    close_arr.append(float(line["close"] ) ) 
    
close_array = np.array(close_arr)
for close in close_array:
    print( close)
print(f"Total records {close_array.size}")
```
