# Array operations with NumPy

In this notebook we are going to learn how to work with array NumPy
- Element wise operations

## Element wise operations

```python
import numpy as np
```

```python
a_array = np.array([1, 2, 3, 4])
b_array = np.array([10, 20, 30, 40])

print(a_array)
print(b_array)
```

```python
# element wise operation
print(a_array + b_array)
print(a_array - b_array)
print(a_array * b_array)
print(a_array / b_array)
print(a_array ** 2)
print(np.sqrt(a_array)) # square root using methos
```

## Broadcasting

Broadcasting allows operations between arrays that have different shapes.
NumPy automatically expands to arrays when possible

NumPy automatically expands arrays when possible

```python
my_array = np.array([1, 2, 3, 4, 5])
print(my_array + 10)
```

Comparison python lists with numpy arrays

```python
import time

 

size = 1_000_0000

 

# Python list

L = list(range(size))

start = time.time()

L2 = [x * 2 for x in L]

print("Python list time:", time.time() - start)

 

# NumPy array

A = np.arange(size)

start = time.time()

A2 = A * 2

print("NumPy array time:", time.time() - start)
```
