### Slicing and Reshaping array

```python
arr = np.arange(1, 13)
print(arr)
```

```python
arr_reshaped = arr.reshape(3,4)
print(arr)
print(arr_reshaped)
```

```python
arr = np.arange(1, 13)
print(arr)
```

```python
reshaped = arr.reshape(4,3)
print(reshaped)
```

```python
extract = reshaped[1:3]
print(extract)
```

```python
# Flatten a multi-dimensional array
flattened_arr = reshaped.ravel()
print(flattened_arr)
```
